```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_pmf(alphas=[0.01, 0.1, 0.5, 1.0], max_i=20):
    x = np.arange(0, max_i + 1)
    plt.figure(figsize=(10, 6))
    for alpha in alphas:
        pmf = poisson.pmf(x, mu=alpha)
        plt.plot(x, pmf, 'o-', label=f'α = {alpha}')
    plt.xlabel('Number of incoming jobs i')
    plt.ylabel('P(X = i)')
    plt.title('Poisson PMF for different arrival rates α')
    plt.legend()
    plt.grid(True)
    plt.savefig('1a_poisson_pmf.png', dpi=200, bbox_inches='tight')
    plt.show()

# Generate samples
def sample_jobs(alpha, size=1):
    return poisson.rvs(mu=alpha, size=size)

plot_poisson_pmf()
```

```python
def simulate_queue(alpha=0.1, proc_time=4, steps=2000, seed=None):
    if seed is not None:
        np.random.seed(seed)
    
    queue = 0
    remaining_proc = 0
    queue_lengths = []
    
    for t in range(steps):
        # Arrivals
        arrivals = sample_jobs(alpha)[0]
        queue += arrivals
        
        # Processing
        if remaining_proc == 0 and queue > 0:
            queue -= 1
            remaining_proc = proc_time
        
        if remaining_proc > 0:
            remaining_proc -= 1
        
        queue_lengths.append(queue)
    
    avg_queue = np.mean(queue_lengths)
    print(f"Average queue length (α={alpha}, proc={proc_time}): {avg_queue:.3f}")
    return avg_queue, queue_lengths

# Run once
avg, _ = simulate_queue(alpha=0.1, proc_time=4, steps=2000)
```

```python
def run_parameter_sweep(proc_time=4, alpha_range=None, runs=200, steps=2000):
    if alpha_range is None:
        alpha_range = np.arange(0.005, 0.251, 0.005) if proc_time == 4 else \
                      np.arange(0.005, 0.501, 0.005)
    
    avg_queues = []
    
    for alpha in alpha_range:
        run_avgs = []
        for r in range(runs):
            avg, _ = simulate_queue(alpha=alpha, proc_time=proc_time, 
                                  steps=steps, seed=r)
            run_avgs.append(avg)
        final_avg = np.mean(run_avgs)
        avg_queues.append(final_avg)
        print(f"α = {alpha:.3f} → avg queue = {final_avg:.3f}")
    
    plt.figure(figsize=(10, 6))
    plt.plot(alpha_range, avg_queues, 'o-', linewidth=2)
    plt.xlabel('Arrival rate α')
    plt.ylabel('Average queue length')
    plt.title(f'Queue length vs α (processing time = {proc_time} steps)')
    plt.grid(True)
    plt.savefig(f'1d_queue_vs_alpha_proc{proc_time}.png', dpi=200, bbox_inches='tight')
    plt.show()
    
    return alpha_range, avg_queues

# Run both experiments
alphas4, queues4 = run_parameter_sweep(proc_time=4)
alphas2, queues2 = run_parameter_sweep(proc_time=2)
```
